```python
import torch
import matplotlib.pyplot as plt
```

```python
snapshot = torch.load('../train_results/test_exp_petals_0/last_snapshot.tar', map_location='cpu')
```

```python
snapshot.keys()
```

```python
history = snapshot['history']
```

```python
history.keys()
```

```python
train_loss = history['train_loss']
```

```python
import numpy as np
```

```python
plt.figure(figsize=(12, 9))
plt.plot(train_loss)
plt.xlabel('epoch')
plt.ylabel('loss')
plt.title('Train loss')
plt.yticks(np.linspace(int(min(train_loss)), int(max(train_loss)), num=10))
plt.grid()
plt.show()
```

```python
snapshot_old = torch.load('../train_results/lm_head_learnable/last_snapshot.tar', map_location='cpu')
history_old = snapshot_old['history']
train_loss_old = history_old['train_loss']

plt.figure(figsize=(12, 9))
plt.plot(train_loss_old)
plt.xlabel('epoch')
plt.ylabel('loss')
plt.title('Train loss')
plt.yticks(np.linspace(int(min(train_loss_old)), int(max(train_loss_old)), num=10))
plt.grid()
plt.show()
```
